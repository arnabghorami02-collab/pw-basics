# Power BI Basics Assignment

 1 : What is Power BI and why is it used in businesses?

Sol :     Power BI is a business analytics and data visualization tool developed by Microsoft. It helps users collect, analyze, and visualize data through interactive dashboards and reports.

Why businesses use Power BI:

- To convert raw data into meaningful insights
- To create interactive reports and dashboards
- To track business performance in real time
- To support better decision-making
- To combine data from multiple sources in one place
- To share reports easily within an organization

 2 : Name and explain the three main components of Power BI.

 Sol : i . Power BI Desktop

- A Windows application used to create reports and dashboards.
- Used for data cleaning, modeling, and visualization.

ii . Power BI Service

- Cloud-based online service.
- Used for publishing, sharing, and collaborating on reports.

iii . Power BI Mobile

- Mobile app for Android and iOS devices.
- Allows users to view reports and dashboards anywhere.


3 : Explain the Power BI workflow.


Sol :

 Power BI Workflow:

- Connect Data
  - Import data from Excel, SQL, web, etc.
- Transform Data
  - Clean and modify data using Power Query.
- Model Data
  - Create relationships between tables and calculations.
- Create Visualizations
  - Build charts, graphs, maps, and dashboards.
- Publish Reports
  - Upload reports to Power BI Service.
- Share & Collaborate
  - Share dashboards with team members.
- Refresh Data
  - Update reports automatically with latest data.

 6 : Define Data View, Report View, and Model View. Explain the purpose of each view.


Sol :     
-  Report View: This is the visual canvas. Purpose: It is used to design and arrange visualizations (charts, maps, tables) to create interactive reports for the end-user.

- Data View: This displays your data in a traditional row-and-column format. Purpose: It allows you to inspect the raw data loaded into your model, verify formatting, and create new calculated columns.

- Model View: This shows a visual diagram of your tables. Purpose: It is used to manage and create relationships (links) between multiple tables in your dataset and understand the overall data architecture.


7 : Discuss the different data sources that Power BI supports.

Sol :     
- Files: Such as Excel, CSV, XML, JSON, and PDF files.

- Databases: Including SQL Server, MySQL, Oracle, PostgreSQL, and IBM DB2.
- Power Platform: Connecting directly to Power BI datasets, Dataflows, and Dataverse.
- Azure: Microsoft cloud services like Azure SQL Database, Azure Blob Storage, and Azure Cosmos DB.
- Online Services: Third-party cloud apps like Salesforce, Google Analytics, GitHub, and SharePoint Online.

- Other: Options to pull data from Web pages (web scraping), OData feeds, Active Directory, or creating blank queries.
